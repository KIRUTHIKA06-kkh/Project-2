Merging Datasets

```python
customer_analysis=pd.merge(sales,customers,on='CustomerKey',how='inner')
```

```python
customer_analysis.dropna(inplace=True)
```

```python
customer_analysis['Delivery Date']=pd.to_datetime(customer_analysis['Delivery Date'])
```

```python
customer_analysis
```

```python
customer_analysis.columns = customer_analysis.columns.str.replace(' ', '')
```

```python
customer_analysis.columns
```

```python
customer_analysis.isnull().sum()
```

```python
customer_analysis.dtypes
```

```python
customer_analysis.to_csv('customer_analysis.csv')
```

```python
sales_analysis=pd.merge(customer_analysis,exchangerates,left_on=['OrderDate'],right_on=['Date'],how='inner')
sales_analysis
```

```python
sales_analysis = pd.concat([sales_analysis, products[['Product Name', 'Unit Cost USD', 'Unit Price USD']]], axis=1)
```

```python
sales_analysis['Product Name'].fillna(value=0, inplace=True)
sales_analysis['Unit Cost USD'].fillna(value=0, inplace=True)
sales_analysis['Unit Price USD'].fillna(value=0, inplace=True)
```

```python
sales_analysis.columns = sales_analysis.columns.str.replace(' ', '')
```

```python
sales_analysis = sales_analysis.loc[:, ~sales_analysis.columns.duplicated()]
```

```python
sales_analysis
```

```python
sales_analysis.columns
```

```python
sales_analysis.isnull().sum()
```

```python
sales_analysis.dtypes
```

```python
sales_analysis.to_csv('sales_analysis.csv')
```

```python
product_analysis=pd.merge(customer_analysis,products,on='ProductKey',how='inner')
product_analysis
```

```python
product_analysis = pd.concat([product_analysis, exchangerates[['Exchange']]], axis=1)
```

```python
product_analysis['CurrencyCode'].fillna(value=0, inplace=True)
product_analysis['Exchange'].fillna(value=0, inplace=True)
```

```python
product_analysis = product_analysis.loc[:, ~product_analysis.columns.duplicated()]
```

```python
product_analysis.columns = product_analysis.columns.str.replace(' ', '')
```

```python
product_analysis.columns
```

```python
product_analysis.isnull().sum()
```

```python
product_analysis.dtypes
```

```python
product_analysis.to_csv('product_analysis.csv')
```
